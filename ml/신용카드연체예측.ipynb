```python
#컬럼명변경
df.rename(columns={'PAY_0':'PAY_1','default.payment.next.month':'default'},inplace=True)
# 속성과 클래스 데이터 분류
y_target = df['default']
X_features= df .drop('default',axis=1)

y_target.value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
#상관관계 

corr=X_features.corr()
corr
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True, fmt='.1g')
# 다중공선성 발생 안좋은 예측결과 도출
# drop 컬럼, 평균을 구해서 사용, 피쳐 셀렉션 방법 중 랜덤 포레스트, 라쏘,릿지를 이용해서 많이 사용한 것을 찾아서 피쳐 사용
# pca를 이용해서 중요한 축 중 주요성부 5개 뽑아서 사용 but 상관도가 높은 피처들 간에는 pca효율이 좋다,
#-> 상관관계 있는 것만 따로 뽑아서 pca해줌
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#bill_amti 1~6까지의 6개의 속성명 생성
cols_bill=['BILL_AMT'+str(i) for i in range(1,7)]
print('대상속성명:',cols_bill)

scaler= StandardScaler()
df_cols_scaled=scaler.fit_transform(X_features[cols_bill])
pca=PCA(n_components=1)
pca.fit (df_cols_scaled)
bills_amt=pca.transform(df_cols_scaled)
print('PCA Component 별 변동성:',pca.explained_variance_ratio_)
bills_amt
```

```python
X_features['BILLS_AMT']=bills_amt
X_features.drop(cols_bill,axis=1, inplace=True)
```

```python
X_features
```

```python

#bill_amti 1~6까지의 6개의 속성명 생성
cols_pay=['PAY_'+str(i) for i in range(1,7)]
print('대상속성명:',cols_pay)

scaler= StandardScaler()
df_cols_scaled=scaler.fit_transform(X_features[cols_pay])
                                    
pca_1=PCA(n_components=1)
pca_1.fit (df_cols_scaled)
pays_amt=pca_1.transform(df_cols_scaled)
print('PCA Component 별 변동성:',pca_1.explained_variance_ratio_)
pays_amt
```

```python
X_features['PAYS_AMT']=pays_amt
X_features.drop(cols_pay, axis=1, inplace=True)
```

```python
X_features
```

```python
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rcf = RandomForestClassifier(n_estimators=300, random_state=156)

scores = cross_val_score(rcf, X_features, y_target, scoring = 'accuracy', cv=3)

print('cv=3개인 경우의 개별 fold세트별 정확도:', scores)
print(f'평균 정확도:{np.mean(scores):.4f}')
```

```python
#pca 변환된 데이터

df_scaled=scaler.fit_transform(X_features)
pca_2=PCA(n_components=7)
df_pca_2=pca_2.fit_transform(df_scaled)
scores_pca_2=cross_val_score(rcf, df_pca_2, y_target, scoring='accuracy', cv=3)

print('cv=3개인 경우의 개별 fold세트별 정확도:', scores_pca_2)
print(f'PCA변환 데이터  셋 평균 정확도:{np.mean(scores_pca_2):.4f}')
```
