```python
import folium
import pandas as pd
```

```python
df = pd.read_csv("./서울특별시강서구안심이CCTV설치현황.csv", 
                encoding = 'cp949')
print("="*80)
print("강서구 안심이 설치현황")
pd.DataFrame(df)
```

```python
for row in df.itertuples():
    print(row)
```

```python

from folium.plugins import MarkerCluster
def san_map(default_location=[37.551111918342, 126.84930138784], 
                default_zoom_start=20):
    base_map = folium.Map(location=default_location, 
                          control_scale=True, 
                          zoom_start=default_zoom_start)
    
    marker_cluster = MarkerCluster().add_to(base_map)

    for row in df.itertuples():
        # print(row.위도)
        if row._3 == '도심공원':
            color='purple'
        elif row._3 == '방범':
            color='green'
        elif row._3 == '불법주차':
            color='pink'
        elif row._3 == '쓰레기 무단투기':
            color='red'
        elif row._3 == '어린이 안전':
            color='blue' 

        folium.Marker([row.위도, row.경도], icon=folium.Icon(color = color), tooltip='도심공원'
                ).add_to(marker_cluster) 


    base_map.save("base_map.html")

        #
# 
# print('='*100)
# print('안심이 Map')
# print('='*100)
san_map()
```
