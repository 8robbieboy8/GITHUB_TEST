```python
import folium

import pandas as pd
```

```python
df = pd.read_csv("./서울특별시강서구안심이CCTV설치현황.csv", 
                encoding = 'cp949')
print("="*80)
print("강서구 안심이 설치현황")
pd.DataFrame(df.head(5))
```

```python
df_sample = df[["안심 주소", 'CCTV 용도', '위도', '경도']]
pd.DataFrame(df_sample.head(5))
```

```python
pd.Series(df_sample.columns)
```

```python
print('데이터 널값 확인')
print(df_sample.isnull().sum())
```

```python
df_drop_sample = df_sample.dropna()
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[37.551111918342, 126.84930138784],
    zoom_start=20
)

coords = df_sample[['위도', '경도']]


marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="pink")).add_to(marker_cluster)


m.save('map.html')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

plt.figure(figsize=(12, 30))
sns.countplot(y=df_sample['CCTV 용도'], order=df_sample['안심 주소'].value_counts().index)
plt.yticks(fontsize=12)
plt.title('강서구 안심이 설치현황')
plt.show()
```

```python
import requests
import json

# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11, 
)

folium.GeoJson(
    seoul_geo,
    name='강서구'
).add_to(m)

m
```

```python
## 서울시 강서구 안심이 설치현황 지도 시각화

gs_map = folium.Map(location = [37.551111918342, 126.84930138784], 
zoom_start=20,
control_scale=True
)

for row in df_sample.itertuples():
    안심주소,	CCTV용도,	위도,	경도 = row[1:]

    if CCTV용도 == '도심공원':
        icon = folium.Icon(color= 'red', icon='info-sign')
    elif CCTV용도 == '방범':
        icon = folium.Icon(color= 'blue', icon='info-sign')
    elif CCTV용도 == '불법주차':
        icon = folium.Icon(color= 'pink', icon='info-sign')
    elif CCTV용도 == '쓰레기 무단투기':
        icon = folium.Icon(color= 'green', icon='info-sign')
    elif CCTV용도 == '어린이 안전':
        icon = folium.Icon(color= 'beige', icon='info-sign')  
    else:
        break  

folium.Marker(location = [위도, 경도],
                popup=f'안심주소 : {안심주소}',
                _icon = icon).add_to(gs_map) 
gs_map
```
