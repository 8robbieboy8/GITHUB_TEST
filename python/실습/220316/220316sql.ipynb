1. emp table 과 dept table을 left join 조건 dept가 같은 경우
2. 출력 컬럼은 ename, job, sal, deptno, loc
3. ename은 그대로 출력
4. job은 각각 한글명으로 변경
5. sal은 2450를 기준으로 2450보다 이상이면 high 미만이면 low
6. deptno은 deptno의 값과 dep table의 dname값을 더해서 표시 (deptno-dname 형식)
7. 모든 컬럼의 이름은 한글로 변경

```python
sql_17 = """
        SELECT emp.ENAME AS "사원명", 
        CASE
            WHEN emp.JOB = "ANALYST" THEN "분석가"
            WHEN emp.JOB = "CLERK" THEN "점원"
            WHEN emp.JOB = "MANAGER" THEN "관리자"
            WHEN emp.JOB = "PRESIDENT" THEN "사장"
            WHEN emp.JOB = "SALESMAN" THEN "판매사원"
            END AS 직업,
        IF(emp.SAL >= 2450, 'HIGH', 'LOW') AS 하이로우,
        emp.SAL AS 판매,
        dept.LOC AS 지역,
        CONCAT(emp.DEPTNO, "-", dept.DEPTNO) AS 부서넘버
        
        FROM emp 
        LEFT JOIN dept
        ON emp.DEPTNO=dept.DEPTNO
        ORDER BY `사원명` ASC
        """
resutl_17 = _db.executeAll(sql_17)
df_17 = pd.DataFrame(resutl_17)
df_17
```
