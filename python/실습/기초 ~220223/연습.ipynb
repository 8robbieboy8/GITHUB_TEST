```python
import pandas as pd

sales_df = pd.read_csv('./220304 pandas/csv/Sales Records.csv')
sales_df.head()
```

```python
sales_df_2 = sales_df.drop(["Ship Date", "Order ID", 
                        "Sales Channel", "Total Revenue"], axis=1)
sales_df_2.head()
```

```python
sales_df_2.columns = ["지역", "국가", "상품타입",
                    "배송순위", "주문날짜", "판매개수",
                    "단가", "원가", "총비용", "총이윤"]

sales_df_2.head()
```

```python
sales_df_2.sort_values('국가', inplace=True)

sales_df_2.head()
```

```python
sales_df_2.reset_index(drop=True, inplace=True)
```

```python
sales_df_2.head()
```

```python
sales_df_2["총이윤"] = sales_df_2["총이윤"].astype(int)
#소수점 제거하고, 정수로 만들기
```

```python
sales_df_2.head()
```

```python
sales_df_2["총비용"] = sales_df_2["총비용"].round(0).astype(int)

sales_df_2.head()
```

```python
sales_df_2["원가"] = sales_df_2["원가"].round(1)
#소수점 하나 남기고 반올림
sales_df_2.head()
```

```python
sales_df_2["판매개수"].round(-1)
```

```python
sales_df_2["주문날짜"] = pd.to_datetime(sales_df_2["주문날짜"],
                        format = "%m/%d/%Y") 
sales_df_2.head()
```

```python
sales_df_2.groupby(["지역", "국가", "상품타입"]).mean()["총이윤"]
```

```python
sales_df_2.groupby(["지역", "국가", "상품타입"]).sum()["총이윤"]
```

```python
sales_df_2.groupby("국가").get_group("South Korea")
```

```python
sales_df_2["지역"].value_counts()
```

```python
sales_df_3 = sales_df_2[sales_df_2['국가'].isin(["South Korea", "China", "Togo"])]
sales_df_3.groupby('국가').mean()['총이윤']
```

```python
import numpy as np

s1 = pd.Series([1, np.nan, 3,4,5])
s2 = pd.Series([1, 2, np.nan, 4, 5])
s3 = pd.Series([1, 2, 3, np.nan, 5])
df = pd.DataFrame({'s1' : s1,
                    's2' : s2,
                    's3' : s3})

df
```

```python
df["s1"].isna()
```

```python
df.dropna(axis=1)
```

```python
df['s1'] = df['s1'] + 1
df
```

```python
_list = [5,4,3,2,1]
df['s2'] = df['s2'] + _list
df
```

```python
_tuple = (1,2,3,4,5)
df['s3'] = df['s3'] + _tuple
df
```

```python
_tuple = (1,2,3,4,5)
df['s3'] += _tuple
df
```

```python
_list = [2,1,2]
dot_result = df @ _list
dot_result
```
